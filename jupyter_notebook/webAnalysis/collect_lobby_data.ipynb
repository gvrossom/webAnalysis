```python
from bs4 import BeautifulSoup

## le fichier est disponible ici: 
# http://ec.europa.eu/transparencyregister/public/consultation/reportControllerPager.do?d-1924860-export=xml&categories=42

with open("lobby_list.xml", "r") as f:
    xml_data = f.read()
    
soup = BeautifulSoup(xml_data, "xml")

rows = soup.find_all('row')[1:]


print "Found %d rows!" % len(rows)

lobby_website_list = []

for row in rows:
    website = row.find_all('column')[6].get_text()
    if website != "":
        lobby_website_list.append(website)
    
print "Extracted %d websites." % len(lobby_website_list)
print "Example of website retrieved:", rows[1].find_all('column')[6].get_text()
```

```python
# lets clean de data to have nice domain names

clean_domains = []
for e in lobby_website_list:
    data = e.split('/')[2]
    if "ww" in data:
        data = data[data.find('.')+1:]
    clean_domains.append(str(data))
    
print "Number of clean domains:", len(clean_domains)
print "Example of clean domain:", clean_domains[1]
```

```python
"eimrail.org", "cer.be" in clean_domains
```

```python
import requests
import json

index_list = ["2016-40"]

def search_domain(domain):

    record_list = []
    
    print "[*] Trying target domain: %s" % domain
    
    for index in index_list:
        
        #print "[*] Trying index %s" % index
        
        cc_url  = "http://index.commoncrawl.org/CC-MAIN-%s-index?" % index
        cc_url += "url=%s&matchType=domain&output=json" % domain
        
        response = requests.get(cc_url)
        
        if response.status_code == 200:
            
            records = response.content.splitlines()
            
            for record in records:
                record_list.append(json.loads(record))
            
            #print "[*] Added %d results." % len(records)
            
    
    print "[*] Found a total of %d hits." % len(record_list)
    
    return record_list

domain_dict = {}

for domain in clean_domains:
    record_list = search_domain(domain)
    print "hits: %d, added to the domain dict." % len(record_list)
    record = ""
    if len(record_list) != 0:
        record = record_list[0]
    domain_dict[domain] = {"hits":len(record_list), "record":record}
```

```python
with open("hits_data.json", "w") as f:
    json.dump(domain_dict, f)
```

```python
import matplotlib.pyplot as plt

hits_vector = []


for domain in clean_domains:
    if domain in domain_dict:
        hits_vector.append(int(domain_dict[domain]["hits"]))

plt.plot(hits_vector)
plt.show()
```

```python
import numpy as np

mean = np.mean(hits_vector)
sd = np.std(hits_vector)

print "Mean:", mean, 
print " Std:", sd

x = mean + sd * hits_vector

n, bins, patches = plt.hist(x, 100, normed=1, facecolor='g', alpha=0.75)


plt.show()
```

```python
short_hits_vector = []
for e in hits_vector:
    if e < 300:
        short_hits_vector.append(e)
print "the vector is now %s's long" % len(short_hits_vector)
```

```python
import numpy as np
new_mean = np.mean(short_hits_vector)
new_std = np.std(short_hits_vector)

print "New Mean:", new_mean,
print "New Std:", new_std

x = new_mean + new_std * short_hits_vector

n, bins, patches = plt.hist(x, 100, normed=2, facecolor='g', alpha=0.75)


plt.grid(True)
plt.show()
```
